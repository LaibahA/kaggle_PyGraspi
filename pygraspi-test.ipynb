# Functions and Generation
The generateGraphAdj() function from igraph_testing is used to generate a graph.

The descriptors() function from decriptors is used to generate a dictionary of t's descriptors.

```python
#This function will create a graph based on any given txt file input
graph_data= ig.generateGraphAdj("test_file.txt")
```

```python
#This function will create a tuple holding in the name and value of all descriptors
dict = d.descriptors(graph_data, "test_file.txt")
```

# Outputting Descriptors
The next couple of lines are basic opening and writing a dictionaries contents to a file.

```python
#We will use the filename below to be the output txt file for the
outputFile = "/kaggle/working/example_descriptors.txt"
```

```python
#Copy descriptor tuple into txt file
with open(outputFile, "w") as f:
    for key, value in dict.items():
        f.write(f"{key}: {value}\n")
```

```python
#Output txt file, if error occurs this means txt file was not properly created
with open(outputFile, 'r') as file:
    content = file.read()
    print(content)
```
